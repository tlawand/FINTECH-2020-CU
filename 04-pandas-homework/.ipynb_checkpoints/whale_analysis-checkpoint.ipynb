# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# set path for csv files
whale_returns_csv = p('./Resources/whale_returns.csv')

# read csv file into dataframe
whale_returns_df = pd.read_csv(whale_returns_csv, parse_dates=True,\
                               infer_datetime_format=True, index_col='Date')

# sort index by ascending order
whale_returns_df.sort_index(ascending=True, inplace=True)

# display few rows from dataframe
whale_returns_df.head()
```

```python
# Count nulls
whale_returns_df.isna().sum()
```

```python
# Drop nulls
whale_returns_df.dropna(inplace=True)
whale_returns_df.head()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = p('./Resources/algo_returns.csv')

# read csv file into dataframe
algo_returns_df = pd.read_csv(algo_returns_csv, parse_dates=True,\
                              infer_datetime_format=True, index_col='Date')

# sort index by ascending order
algo_returns_df.sort_index(ascending=True, inplace=True)

# display few rows from dataframe
algo_returns_df.head()
```

```python
# Count nulls
algo_returns_df.isna().sum()
```

```python
# Drop nulls
algo_returns_df.dropna(inplace=True)
algo_returns_df.head()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices, sorting index
sp500_history_csv = p('./Resources/sp500_history.csv')

# read csv file into dataframe
sp500_history_df = pd.read_csv(sp500_history_csv, parse_dates=True,\
                               infer_datetime_format=True, index_col='Date')

# sort index by ascending order
sp500_history_df.sort_index(ascending=True, inplace=True)

# display few rows from dataframe
sp500_history_df.head()
```

```python
# Check Data Types
sp500_history_df.dtypes
```

```python
# Fix Data Types
sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$','', regex=True)
sp500_history_df['Close'] = sp500_history_df['Close'].astype('float64')
```

```python
# Calculate Daily Returns
sp500_returns = pd.DataFrame(data=sp500_history_df.pct_change(1))
sp500_returns.head()
```

```python
# Drop nulls
sp500_returns.dropna(inplace=True)
sp500_returns.head()
```

```python
# Rename Column
sp500_returns.columns = ['SP500']
sp500_returns.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat([whale_returns_df, algo_returns_df,\
                         sp500_returns], axis='columns', join='inner')
combined_df.head()
```

# Portfolio Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance

Calculate and Plot the daily returns and cumulative returns. 

```python
# Plot daily returns
combined_df.plot(figsize=(20,10))
```

```python
# Plot cumulative returns
cumulative_returns = (1 + combined_df).cumprod() - 1
cumulative_returns.plot(figsize=(20,10))
```

---

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
combined_df.plot(kind='box', figsize=(20,10))
```

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio.
# Which portfolios are riskier than the S&P 500?

portfolio_risk = combined_df.std()
portfolio_risk
```

```python
# Determine which portfolios are riskier than the S&P 500
for i in portfolio_risk.index:
    if portfolio_risk.loc[i] > portfolio_risk['SP500']:
        print(i)
```

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_portfolio_risk = portfolio_risk * np.sqrt(252)
annualized_portfolio_risk
```

---

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&PP 500 using a 21 day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
2. Calculate and plot a 60 day Beta for Berkshire Hathaway Inc compared to the S&&P 500

```python
# Calculate and plot the rolling standard deviation for the S&PP 500 using a 21 day window
rolling_std = combined_df.rolling(window=21).std()
rolling_std.plot(figsize=(20,10))
```

```python
# Correlation
rolling_corr = combined_df.rolling(window=21).corr()
rolling_corr.dropna(inplace=True)
rolling_corr
```

```python
# calculate covariance of berkshire returns to those of sp500
cov_berkshire = combined_df['BERKSHIRE HATHAWAY INC'].\
                rolling(window=21).cov(combined_df['SP500'])
cov_berkshire.dropna(inplace=True)

# calculate variance of sp500
var_sp500 = combined_df['SP500'].rolling(window=21).var()
var_sp500.dropna(inplace=True)

# Calculate Beta for a single portfolio compared to the total market (S&P 500)
beta_berkshire = cov_berkshire / var_sp500
beta_berkshire.dropna(inplace=True)
beta_berkshire
```

### Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the `ewm` with a 21 day half-life.

```python
weighted_returns = combined_df.ewm(span=21).mean()
weighted_returns.plot(figsize=(20,10))
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

Calculate and plot the annualized Sharpe ratios for all portfolios to determine which portfolio has the best performance

```python
# calculate annualized sharpe ratios
annualized_sharpe = (combined_df.mean() * np.sqrt(252))\
                     / (annualized_portfolio_risk)
```

 plot() these sharpe ratios using a barplot.
 On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales?

```python
# Visualize the sharpe ratios as a bar plot
annualized_sharpe.plot(kind='bar', figsize=(20,10))

# Algo 1 porfolio outperforms the S&P500
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
MSFT_path = p('./Resources/MSFT.csv')
MSFT_df = pd.read_csv(MSFT_path, parse_dates=True,\
                      infer_datetime_format=True, index_col='Date')
MSFT_df.index = MSFT_df.index.normalize()
MSFT_df.sort_index(ascending=True)
MSFT_df.columns = ['MSFT_close']
MSFT_df.head()
```

```python
# Read the second stock
PYPL_path = p('./Resources/PYPL.csv')
PYPL_df = pd.read_csv(PYPL_path, parse_dates=True,\
                      infer_datetime_format=True, index_col='Date')
PYPL_df.index = PYPL_df.index.normalize()
PYPL_df.sort_index(ascending=True)
PYPL_df.columns = ['PYPL_close']
PYPL_df.head()
```

```python
# Read the third stock
FISV_path = p('./Resources/FISV.csv')
FISV_df = pd.read_csv(FISV_path, parse_dates=True,\
                      infer_datetime_format=True, index_col='Date')
FISV_df.index = FISV_df.index.normalize()
FISV_df.sort_index(ascending=True)
FISV_df.columns = ['FISV_close']
FISV_df.head()
```

```python
# Read the fourth stock
BRK_B_path = p('./Resources/BRK_B.csv')
BRK_B_df = pd.read_csv(BRK_B_path, parse_dates=True,\
                      infer_datetime_format=True, index_col='Date')
BRK_B_df.index = BRK_B_df.index.normalize()
BRK_B_df.sort_index(ascending=True)
BRK_B_df.columns = ['BRK_B_close']
BRK_B_df.head()
```

```python
# Read the fifth stock
AXP_path = p('./Resources/AXP.csv')
AXP_df = pd.read_csv(AXP_path, parse_dates=True,\
                      infer_datetime_format=True, index_col='Date')
AXP_df.index = AXP_df.index.normalize()
AXP_df.sort_index(ascending=True)
AXP_df.columns = ['AXP_close']
AXP_df.head()
```

```python
# Concatenate all stocks into a single DataFrame
portfolio_df = pd.concat([MSFT_df, PYPL_df, FISV_df, BRK_B_df, AXP_df],\
                         axis='columns', join='inner')
portfolio_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/5, 1/5, 1/5, 1/5, 1/5]

portfolio_returns = portfolio_df.pct_change(1)
portfolio_returns.dropna(inplace=True)

# weighted returns are equal to the annualized returns per stock
# multipllied by their respective weights
weighted_portfolio = portfolio_returns * weights
weighted_portfolio.dropna(inplace=True)
weighted_portfolio
```

```python
weighted_portfolio_annualized = portfolio_returns.mean() * np.sqrt(252) * weights
weighted_portfolio_annualized
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
portfolio_joined = pd.DataFrame(portfolio_returns.sum(axis=1))
portfolio_joined.columns = ['TL']
portfolio_joined
consolidated_returns = pd.concat([combined_df, portfolio_joined], axis='columns', join='inner')
consolidated_returns.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
consolidated_returns.dropna(inplace=True)
consolidated_returns.head()
```

```python
# box plot to show returns per portfolio 
# (showing 10th, 25th, 50th, 75th, and 90th percentiles as well as outliers)
consolidated_returns.plot(kind='box', figsize=(20,10))
```

```python
# rolling beta
tl_rolling_beta = conso_rolling_cov['TL'] / rolling_var_sp500
tl_rolling_beta.plot()
```
